```python
cat_columns = data.select_dtypes(include='object')
real_columns = data.select_dtypes(include='number')
#real_columns = data['id', 'accept_handicapped', 'accept_kids', 'employer_id', 'salary_mean_net']
print(cat_columns)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
import re
import nltk
b = dict()
columns_many = []
for i in cat_columns:
    if len(data[i].unique()) > 30:
        columns_many.append(i)
        for j in range(0, len(data[i])):
            #print(data[i].iloc[j])
            text = re.split('[:/ .()"-01233456789!?;•]+', str(data[i].iloc[j]))
            #print(text)
            for word in text:
                word = word.lower()
                if len(word) > 2:
                    if not word in b: b[word] = 0
                    b[word] = b[word] + 1
arr = []
for i in b:
    arr.append([b[i], i])
arr.sort()
arr.reverse()
c = 0
for i in arr:
    if len(i[1]) > 2:
        #print(i[0], i[1])
        if i[0] >= 4000: c += 1
print(c)
```

```python
arr
```

```python
mp = []
new = []
for i in arr:
    if len(i[1]) > 2:
        if i[0] >= 4000:
            mp.append(i[1])
            new.append([0] * len(data['id']))
for i in cat_columns:
    if len(data[i].unique()) > 30:
        for j in range(0, len(data[i])):
            text = re.split('[:/ .()"-01233456789!?;•]+', str(data[i].iloc[j]))
            for word in text:
                word = word.lower()
                if len(word) > 2:
                    if word in mp:
                        new[mp.index(word)][j] += 1
```

```python
for i in mp:
    data[i] = new[mp.index(i)]
```

```python
new = []

for i in mp:
    new.append([0] * len(test['id']))
            
for i in columns_many:
    for j in range(0, len(test[i])):
        text = re.split('[:/ .()"-01233456789!?;•]+', str(test[i].iloc[j]))
        for word in text:
            word = word.lower()
            if len(word) > 2:
                if word in mp:
                    new[mp.index(word)][j] += 1
```

```python
for i in mp:
    test[i] = new[mp.index(i)]
```

```python
data = data.drop(columns_many, axis = 'columns')
```

```python
test = test.drop(columns_many, axis = 'columns')
```

```python
datatypes = data.dtypes
c = 0
for i in data:
    if i != 'id' and datatypes[c] == 'int64':
        mn = min(data[i].min(), test[i].min())
        mx = max(data[i].max(), test[i].max())
        data[i] = (data[i] - mn) / (mx - mn)
        test[i] = (test[i] - mn) / (mx - mn)
    c += 1
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(data,y,test_size=0.001,random_state=0)
```

```python
from catboost import CatBoostRegressor
from sklearn.model_selection import GridSearchCV

# Определение модели CatBoostRegressor
from sklearn import pipeline
from sklearn import preprocessing
from sklearn.linear_model import LinearRegression,ElasticNet, Lasso, BayesianRidge, Ridge
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import HistGradientBoostingRegressor
from sklearn.datasets import make_hastie_10_2

model = CatBoostRegressor(iterations = 300000)
```

```python
model.fit(X_train, y_train, verbose = 100)
```

```python
# Использование модели с лучшими параметрами для предсказания новых данных
from sklearn.metrics import mean_absolute_percentage_error
y_pred = model.predict(X_test)
mae_rand_forest = mean_absolute_percentage_error(y_test,y_pred)
mae_rand_forest
```
